# 200 DATOS EN PESO Y ALTURAS EN LIBRAS 

###### En el ser humano, la altura y el peso son en su mayoría rasgos hereditarios , pero los factores de estilos de vida, dieta, salud y medio ambiente tambien juegan un papel en la determinación de las caracteristicas fisicas del individuo. El conjunto de datos siguinete contiene 200 registros de alturas y pesos humanos. Estos datos fueron obtenidos en 1993 por un estudio prospectivo de entre 25,000 niños desde el nacimiento hasta los 18 años de edad, los cuales se usarón para elaborar las tablas actuales de peso-altura, edad-peso de Hong-Kong. Se desea obtener un modelo por medio del cual pueda determinarse la altura de una persona al conocer su peso.

## 1.- Generacion del conjunto de Datos

```python
import pandas as pd

import numpy as np
```

```python
x = pd.read_csv("altura_x.csv")
y = pd.read_csv("peso_y.csv")
print("Conjuto de datos en x es: ", len(x))
print("Conjuto de datos en y es: ", len(y))
```

### 2.- Graficacion del conjunto de datos de peso y altura 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x,y,"b.")
plt.show()
```

```python
plt.plot(x,y,"b.")
plt.xlabel("estaturas promedio en libras ")
plt.ylabel("pesos promedio en libras ")
plt.show()
```

### 3.- Modificación del conjunto de  en peso y altura 

```python
import numpy as np
```

```python
data = {'estatura' : x.to_numpy().flatten(), 'peso':y.to_numpy().flatten()}
df = pd.DataFrame(data)
df.head()
```

```python
#Escalado de nUmero de estatura
df['estatura'] = df['estatura'].astype('float')
#Escalado del peso 
df['peso'] = df['peso'].astype('float')
df.head(10)
```

```python
#Representacion grafica del conjunto de datos
plt.plot(df["estatura"],df["peso"],"b.")
plt.xlabel("estatura")
plt.ylabel("peso")
plt.show()
```

### 4.- Construcción del Modelo

```python
from sklearn.linear_model import LinearRegression
```

```python
# constucion del modelo de afustes de hipotesis 

lin_reg = LinearRegression ()
lin_reg.fit(df['estatura'].values.reshape(-1, 1), df['peso'].values)
```

```python
# Parametro theta 0
lin_reg.intercept_
```

```python
# parametro de teta 1
lin_reg.coef_
```

```python
     # Predicción para el valor minimo y Máximo de datos de entrenamiento
x_min_max = np.array([[df['estatura'].min()],[df['estatura'].max()]])
y_train_pred = lin_reg.predict(x_min_max)
```

```python
# Representación gráfica de la función de hipótesis generada
plt.plot(x_min_max, y_train_pred, "g--")
plt.plot(df['estatura'], df['peso'],"b.")
plt.xlabel("estatura")
plt.ylabel("peso")
plt.show()
```

```python
x_new = np.array([[100]]) #1300 equipos afectados

# La predicción del costo que tendria el incidente
Peso = lin_reg.predict(x_new)

print("La altura ", int(Peso[0]))
```

```python
plt.plot(df['estatura'], df['peso'],"b.")
plt.plot(x_min_max, y_train_pred, "g-")
plt.plot(x_new, Peso, "rx")
plt.xlabel("Estatura")
plt.ylabel("Peso")
plt.show()
```
