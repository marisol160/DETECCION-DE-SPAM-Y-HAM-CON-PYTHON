## Representacion Grafica de Datos

Si a la funcion de trasado se le da una matriz de datos, la usa como coordenadas en el eje vertical, utiliara el indice de cada punto de datos en el array como la coordenada horizantal.

```python
plt.plot([1, 2, 5, 7, 3, 1])
plt.show()
```

Tambien se puede proporcionar dos matrices una para el eje horizontal y otra para el eje vertical.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 19]) # Xmin, Xmax, Ymin, Ymax
plt.show()
```

Se sigue el mismo procedimiento para pintar una misma funcion matematica

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Tambien pueden modificarse el estilo de la grafica para que contenga mas informacion.

```python
plt.plot(x, y)
plt.title("Squers Function")
plt.xlabel("X")
plt.ylabel("Y = x**2")
plt.grid(True)
plt.show()
```

Pueden superponerse garficas y cambiar el estilo de las lineas

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es recomendable añadir una leyenda.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best") # La coloca en la localizacion mas chida
plt.show()
```

Tambien puede crearce dos graficas que no se superpongan. Estas graficas se organizab en un grid y se deneminan subplots.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x+1

plt.subplot(1, 2, 1) # 1 Fila, 1 Columna, 1st Subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 Fila, 1 Columna, 2st Subplot
plt.plot(x, y2, 'g')

plt.show()
```

Para que las graficas no queden tan ajustadas, podemos hacer la figura mas grande

```python
plt.figure(figsize = (14,6))

plt.subplot(1, 2, 1) # 1 Fila, 1 Columna, 1st Subplot
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize = 14)
plt.ylabel("y1", fontsize = 14)

plt.subplot(1, 2, 2) # 1 Fila, 1 Columna, 2st Subplot
plt.plot(x, y2, 'g')
plt.xlabel("x2", fontsize = 14)
plt.ylabel("y2", fontsize = 14)

plt.show()
```

### Scatter Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c ='blue')
plt.show()
```

### Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 4]
plt.subplot(211) # 2 rows, 1 columns, 1st subplot
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(212) # 2 rows, 1 columns, 2nd subplot
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel("Value")
plt.ylabel("Frecuency")
plt.show()
```

### Guardar Figuras

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best")
plt.savefig("Mi_grafica_picuda.png", transparent = True)
```
